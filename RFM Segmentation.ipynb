```python
#Dropping NA values as it is not possible to impute them with some random inferences
mydata.dropna(axis = 0,inplace = True)
```

```python
#Converting the date columns to proper format
mydata['First_Date'] = pd.to_datetime(mydata['First_Date'],format = '%Y%m%d')
mydata['Recent_Date'] = pd.to_datetime(mydata['Recent_Date'],format = '%Y%m%d')
```

```python
#Calculating recency data
mydata['recency'] = (mydata['Recent_Date'] - mydata['First_Date']).dt.days
```

```python
#Renaming columns to Recency,Frequency and Monetary
mydata.rename(columns={ 'Buy_Times': 'frequency', 
                         'Customer_Value': 'monetary_value'}, inplace=True)
```

## As the data has monetary value column, we can perform value based segmentation alongside of RFM Segmentation.

# Value Based Segmentation

```python
#creating Deciles of the monetary value column
mydata['decile'] = pd.qcut(mydata['monetary_value'], 10, labels=False)
```

# RFM Segmentation

```python
mydata.head()
```

```python
#Creating RFM Quantiles
quantiles = mydata[['recency' ,'frequency','monetary_value']].quantile(q=[0.25,0.5,0.75])
```

```python
#We create two classes for the RFM segmentation since, being high recency is bad, while high frequency and monetary value is good.
# Arguments (x = value, p = recency, monetary_value, frequency, k = quartiles dict)
def RClass(x,p,d):
    if x <= d[p][0.25]:
        return 1
    elif x <= d[p][0.50]:
        return 2
    elif x <= d[p][0.75]: 
        return 3
    else:
        return 4
    
# Arguments (x = value, p = recency, monetary_value, frequency, k = quartiles dict)
def FMClass(x,p,d):
    if x <= d[p][0.25]:
        return 4
    elif x <= d[p][0.50]:
        return 3
    elif x <= d[p][0.75]: 
        return 2
    else:
        return 1
```

```python
#Creating different quartiles
mydata['R_Quartile'] = mydata['recency'].apply(RClass, args=('recency',quantiles,))
mydata['F_Quartile'] = mydata['frequency'].apply(FMClass, args=('frequency',quantiles,))
mydata['M_Quartile'] = mydata['monetary_value'].apply(FMClass, args=('monetary_value',quantiles,))
```

```python
Final_data = mydata.assign(R_seg=np.nan, FM_seg =np.nan)

#Profiling for Recency
Final_data.R_seg[mydata['R_Quartile']==4] = "3.Churn"
Final_data.R_seg[mydata['R_Quartile']==2] = "2.At Risk"
Final_data.R_seg[mydata['R_Quartile']==3] = "2.At Risk"
Final_data.R_seg[mydata['R_Quartile']==1] = "1.Active"

#Profiling for frequency and monetary
Final_data.FM_seg[(mydata['F_Quartile']==3) & (mydata['M_Quartile']==1)] = "2.Gold"
Final_data.FM_seg[(mydata['F_Quartile']==4) & (mydata['M_Quartile']==1)] = "2.Gold"
Final_data.FM_seg[(mydata['F_Quartile']==2) & (mydata['M_Quartile']==2)] = "2.Gold"
Final_data.FM_seg[(mydata['F_Quartile']==2) & (mydata['M_Quartile']==1)] = "2.Gold"
Final_data.FM_seg[(mydata['F_Quartile']==3) & (mydata['M_Quartile']==2)] = "2.Gold"
Final_data.FM_seg[(mydata['F_Quartile']==4) & (mydata['M_Quartile']==2)] = "1.Silver"
Final_data.FM_seg[(mydata['F_Quartile']==1) & (mydata['M_Quartile']==3)] = "2.Gold"
Final_data.FM_seg[(mydata['F_Quartile']==2) & (mydata['M_Quartile']==3)] = "2.Gold"
Final_data.FM_seg[(mydata['F_Quartile']==3) & (mydata['M_Quartile']==3)] = "1.Silver"
Final_data.FM_seg[(mydata['F_Quartile']==4) & (mydata['M_Quartile']==3)] = "1.Silver"
Final_data.FM_seg[(mydata['F_Quartile']==1) & (mydata['M_Quartile']==4)] = "2.Gold"
Final_data.FM_seg[(mydata['F_Quartile']==1) & (mydata['M_Quartile']==2)] = "2.Gold"
Final_data.FM_seg[(mydata['F_Quartile']==2) & (mydata['M_Quartile']==4)] = "1.Silver"
Final_data.FM_seg[(mydata['F_Quartile']==3) & (mydata['M_Quartile']==4)] = "1.Silver"
Final_data.FM_seg[(mydata['F_Quartile']==4) & (mydata['M_Quartile']==4)] = "0.Standard"

Final_data.FM_seg.fillna("4.Premium", inplace=True)
```
